```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to generate torus points with added noise
def generate_torus(R1, R2, n_points=1000, noise_scale=0.1):
    theta = np.linspace(0, 2 * np.pi, n_points)
    phi = np.linspace(0, 2 * np.pi, n_points)
    theta, phi = np.meshgrid(theta, phi)
    theta = theta.flatten()
    phi = phi.flatten()
    
    # Parametric equations for a torus
    x = (R1 + R2 * np.cos(theta)) * np.cos(phi)
    y = (R1 + R2 * np.cos(theta)) * np.sin(phi)
    z = R2 * np.sin(theta)
    
    # Add Gaussian noise
    x += np.random.normal(0, noise_scale, x.shape)
    y += np.random.normal(0, noise_scale, y.shape)
    z += np.random.normal(0, noise_scale, z.shape)
    
    return x, y, z

# Generate three different tori with different R1 and R2 values
torus1 = generate_torus(R1=5, R2=2)
torus2 = generate_torus(R1=6, R2=1.5)
torus3 = generate_torus(R1=4, R2=3)

# Plotting the three tori in subplots
fig = plt.figure(figsize=(15, 5))

# Plot 1: Torus 1
ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(*torus1, color='blue', alpha=0.7)
ax1.set_title('Torus with R1=5, R2=2')

# Plot 2: Torus 2
ax2 = fig.add_subplot(132, projection='3d')
ax2.scatter(*torus2, color='green', alpha=0.7)
ax2.set_title('Torus with R1=6, R2=1.5')

# Plot 3: Torus 3
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(*torus3, color='red', alpha=0.7)
ax3.set_title('Torus with R1=4, R2=3')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to generate points on a torus
def generate_torus(R1, R2, n_points=1000):
    theta = np.linspace(0, 2 * np.pi, n_points)
    phi = np.linspace(0, 2 * np.pi, n_points)
    theta, phi = np.meshgrid(theta, phi)
    theta = theta.flatten()
    phi = phi.flatten()

    x = (R1 + R2 * np.cos(theta)) * np.cos(phi)
    y = (R1 + R2 * np.cos(theta)) * np.sin(phi)
    z = R2 * np.sin(theta)
    
    return x, y, z

# Parameters you can modify
R1 = 10  # Major radius (distance from the center of the tube to the center of the torus)
R2 = 0.5  # Minor radius (radius of the tube)

# Generate torus points
x, y, z = generate_torus(R1, R2)

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot torus
ax.scatter(x, y, z, c=np.sqrt(x**2 + y**2), cmap='viridis', s=0.05,alpha = 0.1)

# Set axis limits and labels
ax.set_xlim([-R1-R2, R1+R2])
ax.set_ylim([-R1-R2, R1+R2])
ax.set_zlim([-R2, R2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set viewing angle
ax.view_init(elev=30, azim=45)  # You can modify elev and azim to change the view

plt.show()
```
